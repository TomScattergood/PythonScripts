```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
omega_d = np.linspace(0.001, 100, 1000)
b = np.logspace(6, 8, 10)
```

```python
# define constants
F0 = 1e8
m = 660000
k = 1e8
omega_sq = k/m
```

```python
# for handling scientific notation
import matplotlib.ticker as mticker
f = mticker.ScalarFormatter(useMathText=True)
f.set_powerlimits((-3,3))

for b_val in b:
    A = F0/m/((omega_sq-omega_d**2)**2 + (b_val*omega_d/m)**2)**0.5
    
    # formats the b values into a nice string 
    b_string = f._formatSciNotation('1.1e' % b_val)
    
    plt.plot (omega_d, A, label='b=${}$'.format(b_string))
    plt.xlabel(r'$omega_d$ (rad s$^{-1}$)', fontsize=16)
    plt.ylabel('A (m)', fontsize=16)
    
plt.legend
```
